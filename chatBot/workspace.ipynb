## 测试 ConversationBufferWindowMemory
这是一个带记忆窗口的记忆体，为了测试方便，我把的记忆设置成了一条。

```python
from langchain.chains import ConversationChain  
from langchain.memory import ConversationBufferWindowMemory
from langchain.chat_models import ChatOpenAI

chatLLM = ChatOpenAI(temperature=0, model_name="gpt-3.5-turbo",openai_api_base="http://3.101.17.202/v1")

memory = ConversationBufferWindowMemory(k=1)
memory.save_context({"input": "我最喜欢杨幂"}, {"output": "我知道了"})
memory.save_context({"input": "我还喜欢孙燕姿"}, {"output": "我记住了"})

conversation = ConversationChain(llm=chatLLM,memory=memory)  
conversation.run("我最喜欢谁？")
```

### 1. 储存记忆

```python
import time
from langchain.document_loaders import WikipediaLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings


text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
loader = WikipediaLoader(query="杨度",lang='zh')
split_docs = loader.load_and_split(text_splitter)
vectorstore = Chroma("yangdu",persist_directory="../resources/chroma_db", embedding_function=HuggingFaceEmbeddings(model_name='shibing624/text2vec-base-chinese'))
for i in range(len(split_docs)):
    if i%10==0:
        print("[{}] {} / {}".format(time.ctime(),i,len(split_docs)))
    vectorstore.add_documents([split_docs[i]])
```

### 2. 测试效果

```python
import time
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chat_models import ChatOpenAI
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationSummaryMemory

memoryLLM = ChatOpenAI(temperature=0, model_name="gpt-3.5-turbo",openai_api_base="http://3.101.17.202/v1")
chatLLM = ChatOpenAI(temperature=0.5, model_name="gpt-3.5-turbo",openai_api_base="http://3.101.17.202/v1")

embeddings=HuggingFaceEmbeddings(model_name='shibing624/text2vec-base-chinese')
vectorstore = Chroma("yangdu",persist_directory="../resources/chroma_db", embedding_function=embeddings)

memory = ConversationSummaryMemory(llm=memoryLLM,memory_key="chat_history",return_messages=True)
retriever = vectorstore.as_retriever()
qa = ConversationalRetrievalChain.from_llm(chatLLM, retriever=vectorstore.as_retriever(), memory=memory)

qa("杨度的朋友不少啊！")
```
